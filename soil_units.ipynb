```python
from parameters_fields import peat_soilprofiles
import pprint

soils = peat_soilprofiles()
for soil in soils.values():
    soil["pF"]["alpha"] = [a / 100.0 for a in soil["pF"]["alpha"]]
SEC_PER_DAY = 86400.0
for soil in soils.values():
    soil["saturated_conductivity"] = [
        k / SEC_PER_DAY for k in soil["saturated_conductivity"]
    ]
```

```python
# --- write new parameter file ---
outfile = "peat_profiles_modified.py"

with open(outfile, "w") as f:
    f.write("def peat_soilprofiles():\n")
    f.write("    \"\"\"\n")
    f.write("    Peat soil profiles (alpha in 1/cm, Ksat in m/s)\n")
    f.write("    \"\"\"\n")
    f.write("    return ")
    pprint.pprint(soils, stream=f, indent=4, width=120)
    f.write("\n")
```
